# Agora iremos mostrar de forma gráfica:
Foi de extrema importância o site https://matplotlib.org/stable/index.html, que é a sua documentação.

```python
# 
df = base[base["Sport"] == "Gymnastics"]
df1 = df[df["Medal"] == "Gold"]
df2 = df1[df1["Sex"] == "F"]
df2["Age"].mean() # média das idades das mulheres que ganharam ouro e praticaram ginástica
plt.hist(df2["Age"])
plt.xlabel("Idade")
plt.ylabel("Quantidade de mulheres")
plt.title("Quantidade de mulheres por idade a ganhar medalha de Ouro")
plt.show()
```

```python
df = base[base["Medal"] == "Gold"]
df["Age"].mean()
plt.hist(df["Age"])
plt.xlabel("Idade")
plt.ylabel("Quantidade de pessoas")
plt.title("Quantidade de pessoas por idade a ganhar medalha de Ouro")
plt.show()
```

```python
df = base[base["Year"] >= 2010]
x = df.Sex
fig, ax = plt.subplots()
plt.title('Quantidade de pessoas por sexo nos últimos 10 anos')
plt.xlabel('Sexo')
plt.ylabel('Quantidade de pessoas')
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

plt.show()
```

```python
df = base[base["Sex"] == "F"]
plt.style.use("ggplot")
plt.figure(figsize = (10, 6))
df["Year"].hist(bins = 120, ec = "k", alpha = .6, color = "royalblue")
plt.title("Distribuição de mulheres por Olimpíadas)
plt.xlabel("Ano")
plt.ylabel("Quantidade de mulheres")
```

```python
df = base[base["Sex"] == "M"]
plt.style.use("ggplot")
plt.figure(figsize = (10, 6))
df["Year"].hist(bins = 120, ec = "k", alpha = .6, color = "royalblue")
plt.title("Distribuição de homens por Olimpíadas")
plt.xlabel("Ano")
plt.ylabel("Quantidade de homens")
```

```python
# plot:
contagem_cidade = base['City'].value_counts()
plt.figure(figsize=(10, 6))
contagem_cidade.plot(kind='bar')
plt.title('Quantidade de pessoas por cidade')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
contagem_esportes = base['Sport'].value_counts()
plt.figure(figsize=(12, 6))
contagem_esportes.plot(kind='bar')
plt.title('Quantidade de pessoas por esporte')
plt.xlabel('Esporte')
plt.ylabel('Quantidade de pessoas')
plt.show()
```
