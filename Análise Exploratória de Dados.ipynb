# Análise Exploratória de Dados
A análise exploratória de dados é um passo importantíssimo para a criação de projetos, pois leva a estudar características sobre um conjunto de dados.
Alguns de seus benefícios são:
- Investigar dados para conhecer suas características.
- Compreender os dados, identificar padrões e obter insights.
- Detectar erros e anomalias, como valores duplicados e erro na obteção de algum dado.
- Perceber dados alterados para fazer, posteriormente, seu tratamento.

Vale lembrar que:
- Cada tipo de comandos a serem executados dependem do tipo de projeto a ser desenvolvido.

As colunas do nosso dataset são:
- Index (ID do jogador)
- Nome do atleta
- Sexo do atleta
- Idade do atleta
- Time, que representa seu país
- Comitê Nacional Olímpico
- Jogo, que é seu ano com sua temporada
- Ano
- Temporada
- Cidade
- Esporte
- Eventos
- Medalha

```python
# importando as bibliotecas necessárias
import pandas as pd # para manipulação e análise de dados
import matplotlib.pyplot as plt # para os gráficos
import numpy as np # para suportar arrays e matrizes
import seaborn as sns # para gráficos estatísticos e informativos
```

# Agora iremos mostrar de forma gráfica:
Foi de extrema importância o site https://matplotlib.org/stable/index.html, que é a sua documentação.

```python
# plot:
contagem_cidade = base['City'].value_counts()
plt.figure(figsize=(10, 6))
contagem_cidade.plot(kind='bar')
plt.title('Quantidade de pessoas por cidade')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
contagem_esportes = base['Sport'].value_counts()
plt.figure(figsize=(10, 6))
contagem_esportes.plot(kind='bar')
plt.title('Quantidade de pessoas por esporte')
plt.xlabel('Esporte')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
# plot:
#x = base.Age
#fig, ax = plt.subplots()
#ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")
#plt.show()

contagem_idade = base['Age'].value_counts()
plt.figure(figsize=(10, 6))
contagem_idade.plot(kind='bar')
plt.title('Quantidade de pessoas por idade')
plt.xlabel('Idade')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
# plot:
x = base.Sex
fig, ax = plt.subplots()
plt.title('Quantidade de pessoas por sexo')
plt.xlabel('Sexo')
plt.ylabel('Quantidade de pessoas')
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

plt.show()
```

```python
# plot:
contagem_time = base['Medal'].value_counts()
plt.figure(figsize=(10, 5))
contagem_time.plot(kind='bar')
plt.title('Quantidade de pessoas por medalha')
plt.xlabel('Medalhas')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
# plot:
contagem_time = base['Team'].value_counts()
plt.figure(figsize=(10, 4))
contagem_time.plot(kind='bar')
plt.title('Quantidade de pessoas por time')
plt.xlabel('Medalhas')
plt.ylabel('Quantidade de Time')
plt.show()
```

```python
# criando o pairplot
sns.pairplot(base)
```

```python
# matriz de correlação entre variáveis
base.corr()
```

```python
# utilizando heatmap do seaborn para uma matriz de correlação mais visual
sns.heatmap(base.corr())
```

```python
correlation_matrix = base.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação')
plt.show()
```

```python
mask = np.triu(base.corr())
mask
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, mask=mask)
```

```python
mask = np.tril(base.corr())
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, mask=mask)
```
